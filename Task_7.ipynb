```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.a. Write a NumPy program to convert an array to a float type

```python

# Assuming arr is your existing array
arr = np.array([1, 2, 3, 4])

# Convert the array to float type
float_arr = arr.astype(float)

print(float_arr)
```

1.b. Write a NumPy program to add a border (filled with 0's) around an existing array

```python

# Assuming arr is your existing array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Add a border filled with 0's
arr_with_border = np.pad(arr, pad_width=1, mode='constant', constant_values=0)

print(arr_with_border)
```

1.c. Write a NumPy program to convert a list and tuple into arrays


```python

# Converting a list to an array
list_data = [1, 2, 3]
array_from_list = np.array(list_data)

# Converting a tuple to an array
tuple_data = (4, 5, 6)
array_from_tuple = np.array(tuple_data)

print("Array from list:", array_from_list)
print("Array from tuple:", array_from_tuple)
```

1.d. To append values to the end of an array:

```python

# Assuming arr is your existing array
arr = np.array([1, 2, 3])

# Values to append
values_to_append = np.array([4, 5, 6])

# Append values to the end of the array
arr = np.append(arr, values_to_append)

print(arr)
```

2.Write a Pandas program to select all columns, except one given column in a DataFrame.

```python

# Assuming df is your DataFrame and 'column_to_exclude' is the column to exclude
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

column_to_exclude = 'B'

# Select all columns except the given column
selected_columns = df.loc[:, df.columns != column_to_exclude]

print(selected_columns)
```

3.Apply and explore various plotting functions on Pima Indians Diabetes data set for performing the

following:


```python
# Assuming diabetes_df as DataFrame
diabetes_df=pd.read_csv("/content/drive/MyDrive/pima-indians-diabetes.data.csv")
diabetes_df
```

```python
#a. Correlation and scatter plots:

# Display correlation matrix
correlation_matrix = diabetes_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatter plot for two specific columns (e.g., 'Glucose' and 'Insulin')
sns.scatterplot(x='glucose_concentration', y='serum_insulin', data=diabetes_df)
plt.title('Scatter Plot: Glucose vs Insulin')
plt.show()
```

```python
#b. Histograms:

# Histogram for a specific column (e.g., 'Glucose')
plt.hist(diabetes_df['glucose_concentration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Glucose Levels')
plt.xlabel('Glucose Levels')
plt.ylabel('Frequency')
plt.show()
```

```python
#c. Three-dimensional plotting:

from mpl_toolkits.mplot3d import Axes3D

# Assuming 'Glucose', 'Insulin', and 'BMI' are columns in your DataFrame
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(diabetes_df['glucose_concentration'], diabetes_df['serum_insulin'], diabetes_df['bmi'], c='blue', marker='o')

ax.set_xlabel('Glucose')
ax.set_ylabel('Insulin')
ax.set_zlabel('BMI')

plt.title('3D Scatter Plot: Glucose, Insulin, and BMI')
plt.show()
```

```python

# Sample data
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create a DataFrame
df = pd.DataFrame(exam_data, index=labels)

# Select rows where attempts are greater than 2
result_df = df[df['attempts'] > 2]

# Display the result
print("Number of attempts in the examination is greater than 2:")
print(result_df[['name', 'score', 'attempts', 'qualify']])
```

5.Apply and explore various plotting functions on UCI data set for performing the following:


```python
#Reading the CSV file
uci_data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
uci_data
```

```python
#a. Normal values:

# Boxplot for numerical variables to identify outliers
sns.boxplot(data=uci_data)
plt.title('Boxplot of Numerical Variables')
plt.show()
```

```python
#b. Density and contour plots:

# Pairplot for density and scatter plots
sns.pairplot(uci_data, kind='kde', diag_kind='kde')
plt.suptitle('Pairplot with Density Plots')
plt.show()

# Contour plot for two specific numerical columns
sns.kdeplot(data=uci_data, x='age 	', y='Glucose', fill=True, cmap='Blues', levels=20)
plt.title('Contour Plot')
plt.show()
```

```python
#c. Three-dimensional plotting:

from mpl_toolkits.mplot3d import Axes3D

# Assuming 'feature1', 'feature2', and 'feature3' are columns in your DataFrame
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(uci_data['BloodPressure'], uci_data['Glucose'], uci_data['Age'], c='blue', marker='o')

ax.set_xlabel('BloodPressure')
ax.set_ylabel('Glucose')
ax.set_zlabel('Age')

plt.title('3D Scatter Plot: Features 1, 2, and 3')
plt.show()
```
